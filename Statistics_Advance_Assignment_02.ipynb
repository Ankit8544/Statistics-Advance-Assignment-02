Q.No-01    What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Ans :-

The Probability Mass Function (PMF) and Probability Density Function (PDF) are both concepts in probability theory used to describe the distribution of random variables. They provide a way to understand the likelihood of different outcomes occurring for discrete and continuous random variables, respectively.

1. **`Probability Mass Function (PMF) :-`**

The PMF is used for discrete random variables, which take on a finite or countable set of distinct values. The PMF gives the probability of each specific value occurring.

Mathematically, for a discrete random variable X, the PMF is defined as:

$P(X = x) = P(X(x))$

Where:
- $P(X = x)$ is the probability that the random variable X takes on the value x.
- $X(x)$ is the event that the random variable X equals x.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the possible outcomes of a fair six-sided die
outcomes = np.arange(1, 7)

# Calculate the PMF for each outcome (since the die is fair, each outcome has a 1/6 probability)
pmf = np.ones_like(outcomes) / 6

# Plotting the PMF
plt.stem(outcomes, pmf, basefmt=" ")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) of a Fair Six-sided Die")
plt.show()
```

2. **`Probability Density Function (PDF) :-`**

The PDF is used for continuous random variables, which can take on any value within a certain range. The PDF gives the likelihood of the variable falling within a specific range of values.

Mathematically, for a continuous random variable $X$, the PDF is denoted as $f(x)$ and satisfies the following properties:
- $f(x)$ is non-negative for all $x$.
- The area under the curve of $f(x)$ over a certain interval represents the probability of the variable falling within that interval.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define a range of x values
x_values = np.linspace(-5, 5, 1000)

# Calculate the PDF for a standard normal distribution at each x value
pdf_values = norm.pdf(x_values, loc=0, scale=1)

# Plotting the PDF
plt.plot(x_values, pdf_values)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Standard Normal Distribution")
plt.show()
```

Q.N0-02    What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans :-

**`The Cumulative Density Function (CDF)`** is a concept used in statistics and probability theory to describe the probability distribution of a random variable. It provides information about the likelihood that a random variable takes on a value less than or equal to a given value. In simpler terms, the CDF gives you the cumulative probability of a random variable being below a certain value.

**`Mathematically, for a continuous random variable X, the CDF is defined as :`**

$F(x) = P(X \leq x)$

Where:
- $F(x)$ is the CDF of the random variable X.
- $P(X \leq x)$ is the probability that the random variable X takes a value less than or equal to \( x \).

For a discrete random variable, the CDF is the sum of the probabilities of all values less than or equal to the given value.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample dataset (you can replace this with your own data)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate the CDF
sorted_data = np.sort(data)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the CDF
plt.plot(sorted_data, cdf, marker='.', linestyle='none')
plt.xlabel('Data')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Density Function (CDF)')
plt.grid(True)
plt.show()
```

**`The Cumulative Density Function (CDF) is used for several important reasons in statistics and probability :`**

1. **Probability Calculation:** The CDF provides a way to calculate the probability that a random variable falls within a certain range. Specifically, the probability that a random variable \(X\) takes on a value between \(a\) and \(b\) can be calculated as $P(a \leq X \leq b) = F(b) - F(a)$, where $F(x)$ is the CDF of $X$.

2. **Percentiles and Quantiles:** The CDF helps in determining percentiles and quantiles of a distribution. The $p$th percentile is the value below which $p\%$ of the data falls. For example, the median (50th percentile) is the value that divides the data into two equal halves.

3. **Descriptive Statistics:** The CDF provides a comprehensive view of the distribution of a random variable. It shows how the probabilities are distributed across the entire range of values, offering insights into the spread, skewness, and other characteristics of the data.

4. **Comparing Distributions:** When comparing different distributions, the CDF allows you to visually assess how they differ in terms of spread, central tendency, and other statistical properties.

5. **Hypothesis Testing:** The CDF is useful for hypothesis testing and making statistical inferences. It helps in calculating critical values and determining whether observed data is consistent with a particular distribution.

6. **Generating Random Samples:** In some cases, you might want to generate random samples from a specific distribution. The inverse of the CDF (Quantile Function) is used for this purpose, allowing you to transform random numbers into values that follow a specific distribution.

7. **Modeling and Simulation:** The CDF plays a crucial role in modeling and simulating various real-world scenarios involving random variables. It helps in understanding the behavior of random processes and making predictions.

8. **Risk Assessment:** In fields like finance and risk management, the CDF can be used to model and assess the likelihood of different outcomes, helping in decision-making and risk analysis.

9. **Reliability Analysis:** The CDF is often used to analyze the reliability of systems and components. It helps in estimating the probability of failure within a certain time frame.

In summary, the Cumulative Density Function is a fundamental tool in statistics and probability theory that provides a wealth of information about the behavior of random variables, enabling various analyses, comparisons, and predictions.

Q.N0-03    What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

Ans :-

The normal distribution, also known as the Gaussian distribution or bell curve, is a fundamental probability distribution that is commonly used to model various real-world phenomena. It's characterized by its bell-shaped curve and is defined by two parameters: the mean $(μ)$ and the standard deviation $(σ)$.

**`Here are some examples of situations where the normal distribution might be used as a model :-`**

1. **Height of Individuals :** The heights of a large population of individuals often follow a roughly normal distribution. The mean height would be the average height of the population, and the standard deviation would indicate how much the heights vary from the mean.

2. **IQ Scores :** Intelligence quotient (IQ) scores are often modeled using a normal distribution. The mean IQ is typically set at 100, and the standard deviation represents the spread of IQ scores around the mean.

3. **Measurement Errors :** In scientific experiments and measurements, there are often small errors or variations. These errors can often be approximated using a normal distribution, with the mean representing the true value and the standard deviation representing the measurement error.

4. **Financial Markets :** Changes in stock prices or other financial variables often exhibit a pattern resembling a normal distribution. This assumption is fundamental in some financial models, like the Black-Scholes option pricing model.

5. **Test Scores :** Scores on standardized tests, such as the SAT or GRE, are often assumed to follow a normal distribution. The mean represents the average score, and the standard deviation represents how much the scores vary around the mean.

6. **Natural Phenomena :** Many natural processes, such as the distribution of rainfall amounts or the distribution of birth weights, can be approximated by a normal distribution.

**`The parameters of the normal distribution (mean and standard deviation) relate to the shape of the distribution as follows :-`**

1. **Mean $(μ)$ :** The mean determines the center of the distribution. It's the point around which the data tends to cluster. Shifting the mean to the left or right will cause the entire distribution to shift accordingly, while keeping its shape unchanged.

2. **Standard Deviation $(σ)$ :** The standard deviation controls the spread or dispersion of the data. A larger standard deviation results in a wider, more spread-out curve, while a smaller standard deviation leads to a narrower, more tightly clustered curve. A larger standard deviation indicates greater variability in the data points.

In summary, the normal distribution is a versatile model that is used to describe many real-world phenomena. The mean and standard deviation are key parameters that determine the center and spread of the distribution, respectively, and these parameters play a crucial role in shaping the characteristics of the bell curve.

Q.No-05    What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

Ans :-

**`The Bernoulli distribution`** is a discrete probability distribution that models the outcomes of a binary or two-category experiment. It describes a random variable that can take on two possible outcomes, typically referred to as "success" and "failure," each with a certain probability of occurrence. The distribution is named after Jacob Bernoulli, a Swiss mathematician.

Mathematically, if we denote the probability of success as $p$ and the probability of failure as $q = 1 - p$, then `the probability mass function (PMF) of the Bernoulli distribution` can be written as:

$$P(X = x) = \begin{cases}
p, & \text{if } x = 1 \text{ (success)}, \\
q, & \text{if } x = 0 \text{ (failure)}.
\end{cases}$$

Here, $X$ is the random variable representing the outcome, and $x$ can take values of either 0 or 1.

**`Example of Bernoulli distribution :`**

Consider the experiment of flipping a fair coin. Let's say we define "success" as getting heads and "failure" as getting tails. If $p$ is the probability of heads (success), then the probability of tails (failure) is $q = 1 - p$. The outcomes of this experiment follow a Bernoulli distribution.

**`Difference between Bernoulli and Binomial distributions :`**

1. **Number of Trials -**
   - Bernoulli Distribution: It models a single trial or experiment with only two possible outcomes (success or failure).
   - Binomial Distribution: It models the number of successes in a fixed number of independent and identical Bernoulli trials.

2. **Number of Outcomes -**
   - Bernoulli Distribution: It deals with a single random variable representing the outcome of a single trial.
   - Binomial Distribution: It deals with a discrete random variable representing the number of successes in a fixed number of trials.

3. **Probability Mass Function (PMF) -**
   - Bernoulli Distribution: The PMF of Bernoulli distribution involves only two probabilities: $p$ (for success) and $q$ (for failure).
   - Binomial Distribution: The PMF of the binomial distribution involves the probabilities of different numbers of successes in the given fixed number of trials.

4. **Notation -**
   - Bernoulli Distribution: Denoted as $B(p)$, where $p$ is the probability of success.
   - Binomial Distribution: Denoted as $B(n, p)$, where $n$ is the number of trials and $p$ is the probability of success in each trial.

5. **Scope -**
   - Bernoulli Distribution: It's a special case of the binomial distribution where $n = 1$.
   - Binomial Distribution: It's a generalization of the Bernoulli distribution to multiple trials.

In summary, the Bernoulli distribution is a building block for the binomial distribution. The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials, while the Bernoulli distribution is focused on the outcome of a single trial.

Q.No-06    Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

Ans :-

**`To calculate the probability that a randomly selected observation from a normally distributed dataset will be greater than 60, by using the z-score formula and the standard normal distribution table.`**`

**The z-score formula :**

$$z = \frac{x - \mu}{\sigma}$$

Where:
- $ x $ is the value we're interested in (in this case, 60)
- $ \mu $ is the mean of the dataset (50)
- $ \sigma $ is the standard deviation of the dataset (10)

Substituting the values into the formula:

$ z = \frac{60 - 50}{10} = 1 $

Now, we need to find the probability corresponding to this z-score using the standard normal distribution table (also known as the z-table). The z-table provides the cumulative probability of observing a z-score less than or equal to a given value.

Looking up the z-score of 1 in the z-table, we find that the cumulative probability is approximately 0.8413.

Now, since we want the probability of a randomly selected observation being greater than 60, we subtract the cumulative probability from 1:

$ P(X > 60) = 1 - 0.8413 = 0.1587 $

So, the probability that a randomly selected observation will be greater than 60 from this normally distributed dataset is approximately 0.1587 or `15.87%`.

**`To calculate the probability that a randomly selected observation from a normally distributed dataset will be greater than 60, by using statistical functions and distributions, including the normal distribution in python.`**`

```python
import scipy.stats as stats

mean = 50
std_dev = 10
value = 60

# Calculate the Z-score
z_score = (value - mean) / std_dev

# Calculate the probability using the complement of CDF
probability_greater_than_60 = 1 - stats.norm.cdf(z_score)

print("Probability :-", probability_greater_than_60)
```

So, the probability that a randomly selected observation will be greater than 60 from this normally distributed dataset is approximately 0.1587 or `15.87%`.

Q.No-07    Explain uniform Distribution with an example.

Ans :-

**`A uniform distribution`**, also known as `a rectangular distribution`, is a probability distribution in which all outcomes or values within a given range are equally likely to occur. In other words, each value has the same probability of being chosen. It's like rolling a fair six-sided die, where each number from 1 to 6 has an equal chance of appearing.

In a uniform distribution, the probability density function (PDF) is constant within the defined range and zero outside that range. The PDF of a continuous uniform distribution can be represented as:

$f(x) = 1 / (b - a) for a <= x <= b$

$f(x) = 0$ $elsewhere$

Where `a` is the lower bound of the distribution and `b` is the upper bound.

Let's create an example using Python to generate random numbers from a uniform distribution using the `numpy` library:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range [a, b) for the uniform distribution
a = 0
b = 10

# Generate random numbers from the uniform distribution
sample_size = 1000
uniform_samples = np.random.uniform(a, b, sample_size)

# Plotting a histogram to visualize the distribution
plt.hist(uniform_samples, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.show()
```

Q.No-08    What is the z score? State the importance of the z score.

Ans :-

The **`z-score`**, also known as the standard score, is a statistical measure that quantifies how many standard deviations a data point is away from the mean of a distribution. It's a way to standardize and compare values from different distributions by transforming them into a common scale.

The formula for calculating the z-score of a data point **x** in a distribution with mean **μ** and standard deviation **σ** is:

$$ z = \frac{x - μ}{σ} $$

The z-score indicates whether a data point is above the mean (positive z-score) or below the mean (negative z-score), and how far away it is from the mean in terms of standard deviations. A z-score of 0 indicates that the data point is exactly at the mean.

The importance of the z-score lies in several aspects:

1. **`Standardization and Comparison :-`** Z-scores allow you to compare values from different distributions. By transforming values into a standard scale, you can easily compare how unusual or typical a data point is within its distribution.

2. **`Identifying Outliers :-`** Outliers are data points that are significantly different from the rest of the data. Z-scores can help identify outliers by flagging data points with very high or very low z-scores, which indicate that they are far from the mean and might be unusual observations.

3. **`Normal Distribution Analysis :-`** Z-scores are commonly used in the context of normal distributions. In a standard normal distribution (a specific type of normal distribution with mean 0 and standard deviation 1), the z-score directly tells you the percentile rank of a data point. This is useful for assessing the relative position of a data point within a normal distribution.

4. **`Hypothesis Testing :-`** Z-scores are crucial in hypothesis testing, where you compare sample statistics to population parameters. By converting sample means into z-scores, you can determine how likely or unlikely it is for a sample to come from a certain population distribution.

5. **`Data Transformation :-`** Z-scores are sometimes used to transform data to achieve certain properties, such as making data more suitable for specific statistical analyses or modeling techniques that assume normality.

Overall, the z-score is a fundamental tool in statistics that facilitates the analysis, comparison, and interpretation of data across different distributions and contexts. It provides a standardized way to understand the relative position and significance of individual data points within a dataset.

Q.No-09    What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans :-

**`The Central Limit Theorem (CLT)`** is a fundamental concept in statistics and probability theory. It states that, under certain conditions, the distribution of the sample means of a large number of independent and identically distributed random variables will approximate a normal (Gaussian) distribution, regardless of the original distribution of the individual variables. In simpler terms, when you take a sufficiently large number of samples from any population and calculate the mean of each sample, those sample means will tend to follow a bell-shaped normal distribution, even if the original data does not necessarily follow a normal distribution.

    The significance of the Central Limit Theorem is immense, as it has several practical implications :

1. **`Approximation to Normality` :** The CLT enables statisticians to make probabilistic statements about sample means even when the original data might not be normally distributed. This is important because the normal distribution is well-understood and has many convenient mathematical properties.

2. **`Inferential Statistics` :** The CLT forms the basis for many inferential statistical methods, such as hypothesis testing and confidence intervals. These methods allow us to draw conclusions about a population based on a sample, with a quantifiable level of confidence.

3. **`Real-world Applications` :** Many real-world phenomena can be modeled by the sum of a large number of smaller, independent factors. These factors, when averaged, tend to exhibit a normal distribution due to the CLT. This makes the normal distribution a powerful tool for modeling and analyzing various processes and systems, from natural phenomena to financial markets.

4. **`Sampling Techniques` :** The CLT guides how samples should be collected and analyzed. It suggests that even if the underlying population is not normally distributed, using sample means to make inferences is often valid and can provide more stable estimates.

5. **`Statistical Process Control` :** In quality control and process monitoring, the CLT helps establish control limits and thresholds for detecting deviations in processes, since the means of samples taken over time tend to cluster around the true population mean.

6. **`Economics and Social Sciences` :** In fields like economics, sociology, and psychology, researchers often work with sample data to draw conclusions about populations. The CLT helps ensure that the methods used for inference are reliable and accurate.

It's important to note that the Central Limit Theorem has assumptions, such as the requirement for a sufficiently large sample size and the independence of observations. Additionally, the rate at which the distribution of sample means approaches normality can vary based on the characteristics of the original distribution. Nevertheless, the CLT remains a foundational concept in statistics that underpins many of the statistical techniques and analyses used across various disciplines.

Q.No-10    State the assumptions of the Central Limit Theorem.

Ans :-

**`The Central Limit Theorem (CLT)`** is a fundamental statistical concept that provides insights into the distribution of sample means from a population, even when the distribution of the population itself might not be normal. The CLT is based on certain assumptions, which include:

1. **Random Sampling:** The samples are drawn randomly from the population of interest. This means that each observation is independent of the others and is selected without any bias.

2. **Independence:** The observations within each sample are independent of each other. This assumption ensures that the behavior of one observation does not influence the behavior of another.

3. **Finite Variance:** The population from which the samples are drawn must have a finite variance. If the variance is infinite, the CLT may not hold.

4. **Sample Size:** The sample size should be sufficiently large. While there is no fixed threshold for a "large" sample size, a general guideline is that the sample size should ideally be greater than 30. However, for distributions that are strongly skewed or have heavy tails, a larger sample size might be necessary for the CLT to apply effectively.

5. **No Extreme Outliers:** The presence of extreme outliers in the data can potentially affect the validity of the CLT, especially if they significantly impact the mean and variance.

6. **Limited Effect of Sample Size on Population:** The sample size should be relatively small compared to the size of the entire population. This assumption ensures that the act of sampling does not substantially alter the characteristics of the population.

Under these assumptions, the Central Limit Theorem states that regardless of the underlying distribution of the population, the distribution of the sample means will approach a normal distribution as the sample size increases. This property of the CLT is extremely useful in statistical inference and hypothesis testing, as it allows us to make inferences about population parameters based on the distribution of sample means.
